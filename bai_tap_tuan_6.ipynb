```python
#W6A3
def spin(a,b):
    l=[]
    for i in range(len(a)):
        l.append(a[i])
    for i in range(b):
        l.append(l[0])
        l=l[1:len(l)]
    return tuple(l)
a=tuple(input().split())
b=int(input())
print(spin(a,b))
```

```python
#W6A4
l = input().strip()
ans = {}
if l:
    parts = l.split()
    for item in parts:
        if ":" in item:
            k, v = item.split(":", 1)
            k = k.strip()
            v = v.strip()
            if not v:
                continue
            if k in ans:
                ans[k].append(v)
            else:
                ans[k] = [v]
print(ans) 
```

```python
#W6A9
def clean(n):
    ans=""
    for i in n:
        if i!="'" and i!=" " and i!="{" and i!="}":
            ans+=i
    return ans           
a=clean(input())
b=clean(input())
parta=a.split(",")
partb=b.split(",")
l=clean(parta+partb)
ans={}
if l:
    partl=l.split(",")
    for i in partl:
        if ":" in i:
            k,v=i.split(":",1) 
            k=k.strip()
            v=v.strip()
            if not v:
                continue
            if k in ans and v.isdigit() and ans[k].isdigit():
                ans[k]=str(int(ans[k])+int(v))
            else:
                ans[k]=v
print(ans)
```

```python
#W6A10
l=list(map(int,input().split()))
k=int(input())
ans=[]
for i in range(len(l)-1):
    for j in range(i+1,len(l)):
        if l[i]+l[j]==k and (l[i],l[j]) not in ans:
            ans.append((l[i],l[j]))
for i in range(len(ans)-1):
    for j in range(i+1,len(ans)):
        if ans[i]>ans[j]:
            ans[i],ans[j]=ans[j],ans[i]

print(ans)

```

```python
#W6A17
def nhap_mang_vuong():
    while True:
        try:
            n = int(input("Nhập số hàng/cột của mảng vuông (n): "))
            if n <= 0:
                print("Số n phải là số nguyên dương. Vui lòng nhập lại.")
            else:
                break
        except ValueError:
            print("Nhập không hợp lệ. Vui lòng nhập số nguyên.")

    matrix = []
    print(f"Nhập {n} hàng, mỗi hàng {n} số (cách nhau bởi dấu cách):")
    for i in range(n):
        while True:
            try:
                row_str = input(f"Hàng {i+1}: ")
                row = list(map(int, row_str.split()))
                if len(row) != n:
                    print(f"Mỗi hàng phải có {n} số. Vui lòng nhập lại hàng {i+1}.")
                else:
                    matrix.append(row)
                    break
            except ValueError:
                print("Dữ liệu không hợp lệ. Vui lòng nhập các số nguyên cách nhau bởi dấu cách.")
    return matrix, n

def in_duong_cheo(matrix, n):
    duong_cheo_chinh = []
    duong_cheo_phu = []
    
    for i in range(n):
        for j in range(n):
            if i == j:
                duong_cheo_chinh.append(matrix[i][j])
            if i + j == n - 1:
                duong_cheo_phu.append(matrix[i][j])
                
    print("\nMảng vuông bạn đã nhập:")
    for row in matrix:
        print(row)
        
    print("\n--- Kết quả ---")
    print("Đường chéo chính:", duong_cheo_chinh)
    print("Đường chéo phụ:", duong_cheo_phu)
if __name__ == "__main__":
    mang_vuong, size = nhap_mang_vuong()
    in_duong_cheo(mang_vuong, size)

```

```python
#W6A18
try:
    n, m, k = map(int, input("Nhập n, m, k (cách nhau bởi dấu cách): ").split())
except ValueError:
    print("Lỗi: Vui lòng nhập số nguyên hợp lệ.")
    exit()
print(f"Nhập {n} hàng, mỗi hàng gồm {m} số nguyên:")
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    if len(row) != m:
        print(f"Lỗi: Hàng phải có {m} số. Vui lòng nhập lại.")
    matrix.append(row)
if 1 <= k <= m:
    column_index = k - 1
    tong_cot_k = 0
    for i in range(n):
        tong_cot_k += matrix[i][column_index]
    print(f"Tổng các phần tử trên cột {k} là: {tong_cot_k}")
else:
    print(f"Lỗi: Chỉ số cột k = {k} không hợp lệ. (k phải từ 1 đến {m})")
```
