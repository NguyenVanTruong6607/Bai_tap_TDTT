```python
#W8A1
class Cylinder:
    def __init__(self,rad,heg):
        self.rad=rad
        self.heg=heg
    def s(self):
        return 2*3.14*self.rad*self.heg+2*3.14*self.rad**2
    def volum(self):
        return (3.14*self.rad**2)*self.heg
r,h=map(int,input().split())
cylinder=Cylinder(r,h)
area=cylinder.s()
vol=cylinder.volum()
print(f"{area:.2f}")
print(f"{vol:.2f}")
```

```python
#W8A3
class Solution:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def outp(self):
        if self.x==0 and self.y==0:
            return "M là gốc toạ độ, khoảng cách = 0"
        if self.x==0 and self.y!=0:
            return f"M thuộc trục tung, khoảng cách = {self.y:.2f}"
        if self.x!=0 and self.y==0:
            return f"M thuộc trục hoành, khoảng cách = {self.x:.2f}"
        else:
            ans=(self.x**2+self.y**2)**0.5
            return f"MO={(ans):.2f}"
x,y=map(float,input().split())
solu=Solution(x,y)
print(solu.outp())

    
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = []   

    def add_item(self, name, price):
        self.items.append({"name": name, "price": price})

    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                return

    def is_empty(self):
        return len(self.items) == 0

    def total_price(self):
        total = 0
        for item in self.items:
            total += item["price"]
        return total

    def show_items(self):
        if self.is_empty():
            print("Giỏ hàng trống")
            return
        for item in self.items:
            print(item["name"], "-", item["price"])

    def clear_cart(self):
        self.items = []
cart = ShoppingCart()

while True:
    print("1. Thêm sản phẩm")
    print("2. Xóa sản phẩm theo tên")
    print("3. Kiểm tra giỏ hàng rỗng")
    print("4. Tính tổng tiền")
    print("5. Hiển thị giỏ hàng")
    print("6. Xóa toàn bộ giỏ hàng")
    print("7. Thoát")

    choice = input("Chọn (1-7): ")

    if choice == "1":
        name = input("Tên sản phẩm: ")
        price = float(input("Giá: "))
        cart.add_item(name, price)

    elif choice == "2":
        name = input("Tên sản phẩm cần xóa: ")
        cart.remove_item(name)

    elif choice == "3":
        if cart.is_empty():
            print("Giỏ hàng đang trống")
        else:
            print("Giỏ hàng có sản phẩm")

    elif choice == "4":
        print("Tổng tiền:", cart.total_price())

    elif choice == "5":
        cart.show_items()

    elif choice == "6":
        cart.clear_cart()
        print("Đã xóa toàn bộ giỏ hàng")

    elif choice == "7":
        ans = input("Có muốn thoát khỏi chương trình không? ")
        if ans.lower() == "yes":
            break



```

```python
#W8A7
class Fraction:
    def __init__(self,num,den):
        if den < 0:
            num = -num
            den = -den
        self.num=num
        self.den=den
    def __str__(self):
        a,b=self.num,self.den
        a, b = abs(a), abs(b)
        while b != 0:
            a, b = b, a % b
        return f"{int(self.num/a)}/{int(self.den/a)}"
a,b,op,c,d=map(str,input().split())
a,b,c,d=int(a),int(b),int(c),int(d)
if op=="+":
    print(Fraction(d*a+b*c,b*d))
if op=="-":
    print(Fraction(d*a-b*c,b*d))
if op=="*":
    print(Fraction(a*c,b*d))
if op=="/":
    print(Fraction(a*d,b*c))
        

```

```python
#W8A8
class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
    def withdraw(self,draw):
        if draw<=self.balance:
            self.balance-=draw
    def __str__(self):
        return f"{self.balance:.2f}"
owner,balance=input().split()
balance=float(balance)
n=int(input())
bankaccount=BankAccount(owner,balance)

for _ in range(n):
    action,amount=input().split()
    amount=float(amount)
    if action=="DEPOSIT":
        bankaccount.deposit(amount)
    if action=="WITHDRAW":
        bankaccount.withdraw(amount)
print(bankaccount.__str__())


```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, year):
        return sum(1 for b in self.books if b.year == year)
library = Library()

n = int(input().strip())

for _ in range(n):
    line = input().strip()

    if line.startswith("ADD"):
        _, data = line.split(" ", 1)
        title, author, year = data.split(";")
        library.add(Book(title, author, int(year)))

    elif line.startswith("COUNTYEAR"):
        _, year = line.split()
        print(library.find_by_year(int(year)))

    elif line.startswith("COUNT"):
        _, author = line.split(" ", 1)
        print(library.count_by_author(author))

    

        
```

```python
#W8A11
import math

class Point3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def vector_to(self, other):
        return Vector3D(
            other.x - self.x,
            other.y - self.y,
            other.z - self.z
        )


class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def cross(self, other):
        return Vector3D(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )

    def length(self):
        return math.sqrt(self.dot(self))


class Plane:
    def __init__(self, p1, p2, p3):
        v1 = p1.vector_to(p2)
        v2 = p1.vector_to(p3)
        self.normal = v1.cross(v2)

A = Point3D(*map(float, input().split()))
B = Point3D(*map(float, input().split()))
C = Point3D(*map(float, input().split()))
D = Point3D(*map(float, input().split()))

plane1 = Plane(A, B, C)
plane2 = Plane(B, C, D)

n1 = plane1.normal
n2 = plane2.normal

cos_phi = abs(n1.dot(n2)) / (n1.length() * n2.length())
phi = math.degrees(math.acos(cos_phi))

print(f"{phi:.2f}")
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = realPart
        self.imaginaryPart = imaginaryPart

    def multiply(self, other):
        real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imag = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real, imag)



a, b = map(float, input().split())
c, d = map(float, input().split())

z1 = ComplexNumber(a, b)
z2 = ComplexNumber(c, d)


result = z1.multiply(z2)

print(int(result.realPart), int(result.imaginaryPart))




```

```python
#W8A13
class Triagle:
    def __init__(self,egde1,egde2,egde3):
        self.egde1=egde1
        self.egde2=egde2
        self.egde3=egde3
    def getArea(self):
        if (self.egde1+self.egde2)<=self.egde3 or (self.egde3+self.egde2)<=self.egde1 or (self.egde1+self.egde3)<=self.egde2 or (self.egde1-self.egde2)>=self.egde3 or (self.egde1-self.egde3)>=self.egde2 or (self.egde3-self.egde2)>=self.egde1:
            return 'invalid'
        else:
            half=(self.egde1+self.egde2+self.egde3)/2
            return f"{(half*(half-self.egde1)*(half-self.egde2)*(half-self.egde3))**0.5:.2f}"
a,b,c=map(int,input().split())
result=Triagle(a,b,c)
print(result.getArea())
```
