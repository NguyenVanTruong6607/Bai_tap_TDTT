```python
#W11A6
def count(path:str,character:str)->int:
    with open(path,'r') as f:
        n=f.read()
        if character.islower():
            ans=n.count(character)+n.count(character.upper())
        else:
            ans=n.count(character)+n.count(character.lower())
        return ans
```

```python
#W11A13
def moveZeroes (fileName: str) -> list :
    with open(fileName, 'r', encoding='utf-8') as f:
        n_line = f.readline()
        if not n_line: return []
        n = int(n_line.strip())
        nums = [int(x) for x in f.readline().split()]
    non_zeros = [x for x in nums if x != 0]
    zero_count = len(nums) - len(non_zeros)
    result = non_zeros + ([0] * zero_count)
    return result
```

```python
#W11A14
import csv
def averageTime(path: str) -> float:
    with open(path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        total_time = 0.0
        count = 0
        for row in reader:
            try:
                time_val = float(row['Time'])
                total_time += time_val
                count += 1
            except ValueError:
               continue
        if count == 0:
            return 0.0
    return round(total_time / count, 2)
```

```python
#W11A15
import csv
def findMovies(path: str) -> list[str]:
    result_movies = []
    with open(path, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f, skipinitialspace=True)
        for row in reader:
            try:
                movie_name = row['movie']
                description = row['description']
                rating = float(row['rating'])
                if rating > 8.0 and description.lower() != 'boring':
                    result_movies.append(movie_name)
            except ValueError:
                continue
    if not result_movies:
        return ["All are boring!"]
    return result_movies
```

```python
#W11A16
def solve(filename):
    with open(filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]

    i = 0
    while i < len(lines):
        line = lines[i]

        if line == "TYPE: SQUARE":
            side = float(lines[i+1].split(":")[1])
            perimeter = 4 * side
            print(f"{perimeter:.2f}")
            i += 2

        elif line == "TYPE: RECTANGLE":
            parts = lines[i+1].split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)
            print(f"{perimeter:.2f}")
            i += 2

        elif line == "TYPE: CIRCLE":
            radius = float(lines[i+1].split(":")[1])
            perimeter = 2 * 3.14 * radius
            print(f"{perimeter:.2f}")
            i += 2
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        nums = [int(x) for x in f.read().split()]
    n = len(nums)
    if n <= 1:
        return []
    answer = [1] * n
    left_product = 1
    for i in range(n):
        answer[i] = left_product
        left_product *= nums[i]
        right_product = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= right_product
        right_product *= nums[i]
    print(answer)
    return answer
```

```python
#W11A18
def estimatedTime(path: str):
    with open(path, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]

    src_parts = lines[0].split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])
    des_parts = lines[1].split()
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])
    velocity = float(lines[2].split()[1])

    distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5

    time = distance / velocity

    return round(time, 2)
```

```python
#W11A20
def imageSmoother(path: str):
    with open(path, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]

    M = int(lines[0])
    img = []

    for i in range(1, M + 1):
        img.append(list(map(int, lines[i].split())))

    result = [[0] * M for _ in range(M)]

    for i in range(M):
        for j in range(M):
            total = 0
            count = 0

            for x in range(i - 1, i + 2):
                for y in range(j - 1, j + 2):
                    if 0 <= x < M and 0 <= y < M:
                        total += img[x][y]
                        count += 1

            result[i][j] = total // count

    for row in result:
        print(*row)
```
