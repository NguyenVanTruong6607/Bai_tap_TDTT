```python
#W9A1
def merge(a,b):
    if not a:
        return b
    if not b:
        return a
    else:
        if a[0]<=b[0]:
            return [a[0]]+merge(a[1:],b)
        else:
            return [b[0]]+merge(a,b[1:])
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))



```

```python
#W9A2
def count_le(a, n, m, x, i, j):
    if i < 0 or j >= m:
        return 0

    if a[i][j] <= x:
        return (i + 1) + count_le(a, n, m, x, i, j + 1)
    else:
        return count_le(a, n, m, x, i - 1, j)
def kth_smallest(a, n, m, k, low, high):
    if low > high:
        return low

    mid = (low + high) // 2
    cnt = count_le(a, n, m, mid, n - 1, 0)

    if cnt >= k:
        return kth_smallest(a, n, m, k, low, mid - 1)
    else:
        return kth_smallest(a, n, m, k, mid + 1, high)
n, m = map(int, input().split())
a = []
for _ in range(n):
    a.append(list(map(int, input().split())))

k = int(input())
print(kth_smallest(a, n, m, k, a[0][0], a[n-1][m-1]))


        

    
```

```python

#W9A3
def min_after_one_swap(n):
    s = str(n)
    L = len(s)
    best = [None]  

    def dfs(i):
        if i == L:
            return

        def dfs2(j):
            if j == L:
                return

          
            lst = list(s)
            lst[i], lst[j] = lst[j], lst[i]

            if lst[0] != '0':
                val = int(''.join(lst))
                if best[0] is None or val < best[0]:
                    best[0] = val

            dfs2(j + 1)

        dfs2(i + 1)
        dfs(i + 1)

    dfs(0)
    return best[0]


n = int(input().strip())
print(min_after_one_swap(n))



```

```python
#W9A4
def check(a, b):
    if not a:
        return not b

    if not b:
        return a.islower()

    if a[0] == b[0]:
        return check(a[1:], b[1:])

    if a[0].islower():
        if a[0].upper() == b[0]:
            return check(a[1:], b[1:]) or check(a[1:], b)
        return check(a[1:], b)

    return False
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print("YES" if check(a, b) else "NO")
    
```

```python
#W9A6
def find_i(a, i):
    if i < 0:
        return -1
    if a[i] < a[i+1]:
        return i
    return find_i(a, i-1)

def find_j(a, i, j):
    if a[j] > a[i]:
        return j
    return find_j(a, i, j-1)

def reverse_rec(a, l, r):
    if l >= r:
        return
    a[l], a[r] = a[r], a[l]
    reverse_rec(a, l+1, r-1)

def next_permutation(a):
    n = len(a)
    i = find_i(a, n-2)
    if i == -1:
        reverse_rec(a, 0, n-1)
        return a
    j = find_j(a, i, n-1)
    a[i], a[j] = a[j], a[i]
    reverse_rec(a, i+1, n-1)
    return a



```

```python
#W9A7
def canPlaceFlowers(flowerbed, k, i=0):
    if k == 0:
        return True
    if i >= len(flowerbed):
        return False

    if flowerbed[i] == 0:
        left = (i == 0 or flowerbed[i-1] == 0)
        right = (i == len(flowerbed)-1 or flowerbed[i+1] == 0)
        if left and right:
            flowerbed[i] = 1
            return canPlaceFlowers(flowerbed, k-1, i+2)

    return canPlaceFlowers(flowerbed, k, i+1)
```

```python
#W9A8
def is_path_crossing(moves, i=0, x=0, y=0, visited=None):
    if visited is None:
        visited = {(0, 0)}

    if i == len(moves):
        return False

    if moves[i] == 'R': x += 1
    elif moves[i] == 'L': x -= 1
    elif moves[i] == 'U': y += 1
    else: y -= 1

    if (x, y) in visited:
        return True

    visited.add((x, y))
    return is_path_crossing(moves, i+1, x, y, visited)
```

```python
#W9A9
def kth_smallest(a, k):
    if len(a) == 1:
        return a[0]

    pivot = a[0]
    left = [x for x in a[1:] if x < pivot]
    right = [x for x in a[1:] if x >= pivot]

    if k <= len(left):
        return kth_smallest(left, k)
    elif k == len(left) + 1:
        return pivot
    else:
        return kth_smallest(right, k - len(left) - 1)

def Median(nums):
    return kth_smallest(nums, 3)
```

```python
#W9A10
def nearest_heater(h, heaters, i=0, best=10**18):
    if i == len(heaters):
        return best
    return nearest_heater(
        h, heaters, i+1,
        min(best, abs(h - heaters[i]))
    )

def findRadius(houses, heaters, i=0, ans=0):
    if i == len(houses):
        return ans
    d = nearest_heater(houses[i], heaters)
    return findRadius(houses, heaters, i+1, max(ans, d))
```
