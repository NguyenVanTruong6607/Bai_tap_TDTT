```python
#W5A5
def index_of_k(arr, k) -> int:
  for i in range(len(arr)):
    if arr[i]==k:
      return i+1
  return -1
if __name__ ==  "__main__":
    arr = list(map(int, input().split()))
    k = int(input())
    print (index_of_k(arr, k))
```

```python
#W5A6
def fac(n):
  if n==0:
    return 1
  ans=1
  for i in range(1,n+1):
    ans*=i
  return ans
n=int(input())
print(fac(n))
```

```python
#W5A7
def ans(a,b,n):
    if n=="+":
        return f"{a+b:.2f}"
    if n=="-":
        return f"{a-b:.2f}"
    if n=="*":
        return f"{a*b:.2f}"
    if n=="/":
        return f"{a/b:.2f}"
a,b=map(float,input("Nhập 2 số cách nhau bắng dấu cách: ").split())
n=str(input("Nhập dấu:"))
print(ans(a,b,n))
```

```python
#W5A8
def hamming(a,b):
    return bin(a^b).count("1")

a,b=map(int,input().split())
print(hamming(a,b))
```

```python
#W5A9
def sum(n):
  ans=0
  while n!=0:
    ans+=(n%10)
    n=n//10
  return ans
n=int(input())
print(sum(n))
```

```python
#W5A10
def check(a,b):
  for i in range(len(a)-1):
    for j in range(i+1,len(a)):
      if a[i]==a[j] and b[i]!=b[j]:
        return False
      if a[i]!=a[j] and b[i]==b[j]:
        return False
  return True
a,b=map(str,input().split())
print(check(a,b))
```
