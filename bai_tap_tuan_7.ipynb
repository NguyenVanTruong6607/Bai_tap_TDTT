```python
#W7A3
    #BUBBLE SORT
def bubble(a:list):
    for i in range(len(a)):
        swap=False
        for j in range(len(a)-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
                swap=True
        if not swap:
            break
    return a


a=list(map(int,input().split()))
print(bubble(a))
```

```python
#W7A3
    #SELECTION SORT
def select(a):
    n=len(a)
    for i in range(n):
        min_idx=i
        for j in range(i+1,n):
            if a[j]<a[min_idx]:
                min_idx=j
                a[i],a[min_idx]=a[min_idx],a[i]
    return a


    
a=list(map(int,input().split()))
print(select(a))

```

```python
#W7A5
def count(a:list,b:int):
    ans=0
    for i in range(len(a)-1):
        for j in range(i+1,len(a)):
            if b==(a[i]+a[j]):
                ans+=1
    return ans
a=list(map(int,input().split()))
b=int(input())
print(count(a,b))


    


```

```python
#W7A9
def find(a:list):
    ans=[]
    k=0
    for i in range(len(a)-1):
        for j in range(i+1,len(a)):
            if a[j]<a[i]:
                k+=1
        ans.append(k)
        k=0
    ans.append(0)
    return ans

a=input()
parts=[]
for i in a:
    if i.isdigit():
        parts.append(int(i))

print(find(parts))
```

```python
#W7A10
def find(a:list,b:int):
    ans=a[0]%b
    for i in range(len(a)):
        sum=a[i]
        for j in range(1,len(a)):
            if ans<((sum+a[j])%b):
                sum+=a[j]
                ans=(sum)%b
            else :  
                sum+=a[j]
    return ans





a=list(map(int,input().split()))
b=int(input())
print(find(a,b))
```

```python
#W7A14
def find_seven(a:list):
    if 7 not in a:
        return "Not found"
    else:
        ans=""
        for i in range(len(a)-1,-1,-1):
            if 7==a[i]:
                ans+=f"{i}"+" "
        return ans


a=int(input())
b=list(map(int,input().split()))
while len(b)!=a:
    b=list(map(int,input("Nhập sai nhập lại").split()))

print(find_seven(b))
```
