```python
#W8A3
try:
    def HappyNumbers(a):
        
        sum=0
        while a!=0:
            sum+=(a%10)**2
            a=a//10
        if sum==1:
            return True
        else: 
            return HappyNumbers(sum)
        
    a=int(input())
    if HappyNumbers(a):
        print("YES")

except ValueError:
    print("not a number")
except RecursionError:
    print("NO")
```

```python
#W8A11
repeat = {'*', '+', '?'}

def is_valid(s):
    for i in range(len(s) - 1):
        if s[i] in repeat and s[i + 1] in repeat:
            return False
    return True


t = int(input())
for _ in range(t):
    s = input().strip()
    print("True" if is_valid(s) else "False")
```
